```python
# Fungsi aktivasi sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Fungsi turunan dari sigmoid
def sigmoid_derivative(x):
    return x * (1 - x)


import numpy as np

class NeuralNetwork:
    def __init__(self, input_size, hidden_size, output_size):
        # Menginisialisasi bobot dari input ke hidden
        self.weights_input_hidden = np.array([[0.5, 0.2, -0.3],
                                              [0.7, -0.4, 0.6]])

        # Menginisialisasi bobot dari hidden ke output
        self.weights_hidden_output = np.array([[0.1, -0.2],
                                               [0.4, 0.3],
                                               [-0.5, 0.7]])

        # Menginisialisasi bias untuk layer hidden
        self.bias_hidden = np.array([0.5, 0.5, 0.5])

        # Menginisialisasi bias untuk layer output
        self.bias_output = np.array([0.5, -0.5])
```

```python
def forward(self, inputs):
    # Forward propagation
    # Menghitung input untuk layer hidden
    self.hidden_input = np.dot(inputs, self.weights_input_hidden) + self.bias_hidden

    # Menghitung output dari layer hidden menggunakan fungsi aktivasi sigmoid
    self.hidden_output = sigmoid(self.hidden_input)

    # Menghitung input untuk layer output
    self.output_input = np.dot(self.hidden_output, self.weights_hidden_output) + self.bias_output

    # Menghitung output akhir menggunakan fungsi aktivasi sigmoid
    self.predicted_output = sigmoid(self.output_input)

    return self.predicted_output
```

```python
def backward(self, inputs, target, learning_rate):
    # Backpropagation
    # Menghitung error pada output
    error = target - self.predicted_output

    # Menghitung delta output menggunakan turunan sigmoid
    delta_output = error * sigmoid_derivative(self.predicted_output)

    # Menghitung error pada layer hidden
    error_hidden = delta_output.dot(self.weights_hidden_output.T)

    # Menghitung delta hidden menggunakan turunan sigmoid
    delta_hidden = error_hidden * sigmoid_derivative(self.hidden_output)

    # Update bobot dari hidden ke output
    self.weights_hidden_output += np.outer(self.hidden_output, delta_output) * learning_rate

    # Update bobot dari input ke hidden
    self.weights_input_hidden += np.outer(inputs, delta_hidden) * learning_rate
```

```python
def train(self, training_data, targets, epochs, learning_rate):
    for epoch in range(epochs):
        total_error = 0  # Inisialisasi total error untuk setiap epoch

        # Loop melalui setiap data dalam training set
        for i in range(len(training_data)):
            inputs = training_data[i]
            target = targets[i]

            # Lakukan forward propagation
            self.forward(inputs)

            # Lakukan backward propagation
            self.backward(inputs, target, learning_rate)

            # Hitung squared error dan tambahkan ke total error
            total_error += np.mean(np.square(target - self.predicted_output))

        # Cetak error setiap 100 iterasi
        if epoch % 100 == 0:
            print(f"Epoch {epoch}, Average Error: {total_error / len(training_data)}")
```

```python
def predict(self, inputs):
    return self.forward(inputs)

# Input atau nilai x nya
training_data = np.array([[0, 5]])
targets = np.array([0])

input_size = 2
hidden_size = 3
output_size = 1
learning_rate = 0.1
epochs = 100

# Inisialisasi jaringan neural
nn = NeuralNetwork(input_size, hidden_size, output_size)

# Melatih jaringan neural
nn.train(training_data, targets, epochs, learning_rate)

# Test network
for inputs in training_data:
    prediction = nn.predict(inputs)
    print(f"Input: {inputs}, Predicted Output: {prediction}")
```

```python
import numpy as np

# Fungsi aktivasi sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Fungsi turunan dari sigmoid
def sigmoid_derivative(x):
    return x * (1 - x)

class NeuralNetwork:
    def __init__(self, input_size, hidden_size, output_size):
        # Menginisialisasi bobot dari input ke hidden
        self.weights_input_hidden = np.random.rand(input_size, hidden_size)

        # Menginisialisasi bobot dari hidden ke output
        self.weights_hidden_output = np.random.rand(hidden_size, output_size)

        # Menginisialisasi bias
        self.bias_hidden = np.random.rand(hidden_size)
        self.bias_output = np.random.rand(output_size)

    def forward(self, inputs):
        # Forward propagation
        self.hidden_input = np.dot(inputs, self.weights_input_hidden) + self.bias_hidden
        self.hidden_output = sigmoid(self.hidden_input)
        self.output_input = np.dot(self.hidden_output, self.weights_hidden_output) + self.bias_output
        self.predicted_output = sigmoid(self.output_input)
        return self.predicted_output

    def backward(self, inputs, target, learning_rate):
        # Backpropagation
        error = target - self.predicted_output
        delta_output = error * sigmoid_derivative(self.predicted_output)

        error_hidden = delta_output.dot(self.weights_hidden_output.T)
        delta_hidden = error_hidden * sigmoid_derivative(self.hidden_output)

        # Update weights
        self.weights_hidden_output += np.outer(self.hidden_output, delta_output) * learning_rate
        self.weights_input_hidden += np.outer(inputs, delta_hidden) * learning_rate

    def train(self, training_data, targets, epochs, learning_rate):
        for epoch in range(epochs):
            total_error = 0  # Inisialisasi total error untuk setiap epoch

            # Loop melalui setiap data dalam training set
            for i in range(len(training_data)):
                inputs = training_data[i]
                target = targets[i]

                # Lakukan forward propagation
                self.forward(inputs)

                # Lakukan backward propagation
                self.backward(inputs, target, learning_rate)

                # Hitung squared error dan tambahkan ke total error
                total_error += np.mean(np.square(target - self.predicted_output))

            # Cetak error setiap 100 iterasi
            if epoch % 100 == 0:
                print(f"Epoch {epoch}, Average Error: {total_error / len(training_data)}")

    def predict(self, inputs):
        return self.forward(inputs)


# Input atau nilai x nya
training_data = np.array([[0, 5]])
targets = np.array([0])

input_size = 2
hidden_size = 3
output_size = 1
learning_rate = 0.1
epochs = 1000  # Perpanjang epochs untuk melihat hasil lebih jelas

# Inisialisasi jaringan neural
nn = NeuralNetwork(input_size, hidden_size, output_size)

# Melatih jaringan neural
nn.train(training_data, targets, epochs, learning_rate)

# Test network
for inputs in training_data:
    prediction = nn.predict(inputs)
    print(f"Input: {inputs}, Predicted Output: {prediction}")
```
