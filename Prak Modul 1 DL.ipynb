# Modul Praktikum 1 DEEP LEARNING

menginisiasi nilai input x, bobot W(1), W(2), dan W(3).

```python
import numpy as np

# Masukkan nilai input x dan bobot W1, W2, dan W3
x = np.matrix([[3], [1], [5]])

w1 = np.matrix([
    [-2, 1, 1, -4],
    [3, 4, 2, -3],
    [1, -1, 1, 3]
])

w2 = np.matrix([
    [-2, 1, -3, -4],
    [3, 2, 1, -1],
    [2, 4, 1, -3],
    [-1, -1, -3, 2]
])

w3 = np.matrix([
    [5, 3, -1],
    [4, 1, -2],
    [1, 3, 1],
    [-2, 4, 3]
])
```

Setelah melakukan inisiasi bobot, selanjutnya adalah inisiasi nilai bias b1, b2, dan b3.

```python
# masukkan nilai bias b1, b2, dan b3
b1 = np.matrix([[-4], [7], [5], [3]])
b2 = np.matrix([[-1],[-2], [1], [5]])
b3 = np.matrix([[4], [-3], [2]])
```

transpose terhadap bobot W(1), W(2)
, dan W(3)

```python
# transpose matriks W1, W2, dan W3
w1_T = np.transpose(w1)
w2_T = np.transpose(w2)
w3_T = np.transpose(w3)
```

proses perhitungan h1 dengan menjumlahkan
hasil perkalian matriks W(1) dengan nilai input x yang kemudian ditambahkan nilai bias
b1.

```python
# hitung nilai dari input ke hidden layer h_1
h_1=(w1_T*x) + b1
```

mengaktifkan fungsi aktivasi sigmoid pada keluaran h1

```python
# Fungsi aktivasi yang bisa digunakan
def sigmoid(mat):
    return 1 / (1 + np.exp(-mat))

def relu(mat):
    return np.maximum(0, mat)

def tanh(mat):
    return np.tanh(mat)

def softmax(mat):
    e_z = np.exp(mat - np.max(mat))
    return e_z / e_z.sum(axis=0)
```

```python
# aktifkan fungsi aktivasi sigmoid pada h_1
h1_aktivasi = sigmoid(h_1)
```

```python
# hitung nilai pada hidden layer h_2
h_2= (w2_T*h1_aktivasi) + b2
```

```python
# aktifkan fungsi aktivasi sigmoid pada h_2
h2_aktivasi = sigmoid(h_2)
```

```python
# # aktifkan fungsi aktivasi ReLU pada output
out = (w3_T*h2_aktivasi) + b3
output = relu(out)

# nilai output
print(output)
```
